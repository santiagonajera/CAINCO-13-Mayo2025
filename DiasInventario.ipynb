```python
import pandas as pd
import numpy as np

# URL del archivo de ventas
url_ventas = "https://github.com/santiagonajera/CAINCO-13-Mayo2025/raw/refs/heads/main/ventas_items_cainco.xlsx "

# URL del archivo de LeadTime
url_leadtime = "https://github.com/santiagonajera/CAINCO-13-Mayo2025/raw/refs/heads/main/Cainco-Costos-LeadTime.xlsx "

# Leer datos de ventas
df_ventas = pd.read_excel(url_ventas)
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])

# Leer datos de LeadTime
df_leadtime = pd.read_excel(url_leadtime, sheet_name='LeadTime')

# Calcular promedio y desviaci칩n est치ndar para cada 칤tem en los datos de ventas
estadisticas_ventas = {}
for item in df_ventas.columns[1:]:
    valores = df_ventas[item]
    promedio = np.mean(valores)
    std = np.std(valores)
    estadisticas_ventas[item] = {
        'Promedio_Ventas': promedio,
        'Desviaci칩n_Est치ndar_Ventas': std
    }

# Calcular promedio y desviaci칩n est치ndar para cada 칤tem en los datos de LeadTime
estadisticas_leadtime = {}
for item in df_leadtime['Item']:
    item_data = df_leadtime[df_leadtime['Item'] == item]
    lead_times = item_data[['LeadTime_1', 'LeadTime_2']].values.flatten()
    promedio = np.mean(lead_times) / 7  # Convertir a semanas
    std = np.std(lead_times)
    estadisticas_leadtime[item] = {
        'Promedio_LeadTime': promedio,
        'Desviaci칩n_Est치ndar_LeadTime': std
    }

# Par치metros adicionales
Z_alpha = 1.645  # Para un nivel de confianza del 95%
R = 0.5 / 7  # Reorden point en semanas

# Calcular el stock de seguridad para cada 칤tem
stocks_seguridad = {}

for item in estadisticas_ventas.keys():
    # Datos de ventas
    mu_d = estadisticas_ventas[item]['Promedio_Ventas']
    sigma_d = estadisticas_ventas[item]['Desviaci칩n_Est치ndar_Ventas']

    # Datos de LeadTime
    mu_L = estadisticas_leadtime.get(item, {}).get('Promedio_LeadTime', 0)
    sigma_L = estadisticas_leadtime.get(item, {}).get('Desviaci칩n_Est치ndar_LeadTime', 0)

    # F칩rmula para calcular el stock de seguridad
    ss = Z_alpha * np.sqrt(
        (mu_L + R) * sigma_d**2 +
        sigma_L**2 * mu_d**2
    )

    stocks_seguridad[item] = ss

# Convertir el diccionario de resultados en un DataFrame
resultados_stock_seguridad = pd.DataFrame(stocks_seguridad.items(), columns=['Item', 'Stock_Seguridad'])
resultados_stock_seguridad.set_index('Item', inplace=True)

# Mostrar los resultados
print("\n游늵 Stock de Seguridad por 칤tem:")
print(resultados_stock_seguridad.round(2))
```

```python
import pandas as pd
import numpy as np

# URL del archivo de ventas
url_ventas = "https://github.com/santiagonajera/CAINCO-13-Mayo2025/raw/refs/heads/main/ventas_items_cainco.xlsx "

# URL del archivo de LeadTime
url_leadtime = "https://github.com/santiagonajera/CAINCO-13-Mayo2025/raw/refs/heads/main/Cainco-Costos-LeadTime.xlsx "

# Leer datos de ventas
df_ventas = pd.read_excel(url_ventas)
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])

# Leer datos de LeadTime
df_leadtime = pd.read_excel(url_leadtime, sheet_name='LeadTime')

# Calcular promedio y desviaci칩n est치ndar para cada 칤tem en los datos de ventas
estadisticas_ventas = {}
for item in df_ventas.columns[1:]:
    valores = df_ventas[item]
    promedio = np.mean(valores)
    std = np.std(valores)
    estadisticas_ventas[item] = {
        'Promedio_Ventas': promedio,
        'Desviaci칩n_Est치ndar_Ventas': std
    }

# Calcular promedio y desviaci칩n est치ndar para cada 칤tem en los datos de LeadTime
estadisticas_leadtime = {}
for item in df_leadtime['Item']:
    item_data = df_leadtime[df_leadtime['Item'] == item]
    lead_times = item_data[['LeadTime_1', 'LeadTime_2']].values.flatten()
    promedio = np.mean(lead_times) / 7  # Convertir a semanas
    std = np.std(lead_times)
    estadisticas_leadtime[item] = {
        'Promedio_LeadTime': promedio,
        'Desviaci칩n_Est치ndar_LeadTime': std
    }

# Par치metros adicionales
Z_alpha = 1.645  # Para un nivel de confianza del 95%
R = 0.5 / 7  # Reorden point en semanas

# Calcular el stock de seguridad para cada 칤tem
stocks_seguridad = {}
stocks_seguridad_dias = {}
techo_inventario_dias = {}
stocks_seguridad_optimo_dias = {}

for item in estadisticas_ventas.keys():
    # Datos de ventas
    mu_d = estadisticas_ventas[item]['Promedio_Ventas']
    sigma_d = estadisticas_ventas[item]['Desviaci칩n_Est치ndar_Ventas']

    # Datos de LeadTime
    mu_L = estadisticas_leadtime.get(item, {}).get('Promedio_LeadTime', 0)
    sigma_L = estadisticas_leadtime.get(item, {}).get('Desviaci칩n_Est치ndar_LeadTime', 0)

    # F칩rmula para calcular el stock de seguridad
    ss = Z_alpha * np.sqrt(
        (mu_L + R) * sigma_d**2 +
        sigma_L**2 * mu_d**2
    )

    # Stock de seguridad en d칤as
    ss_dias = (ss / mu_d) * 7

    # Techo de inventario en d칤as
    techo_inventario = ss_dias + (mu_L * 7)

    # Stock de seguridad 칩ptimo en d칤as
    ss_optimo_dias = ss_dias + ((techo_inventario - ss_dias) / 2)

    # Guardar resultados
    stocks_seguridad[item] = ss
    stocks_seguridad_dias[item] = ss_dias
    techo_inventario_dias[item] = techo_inventario
    stocks_seguridad_optimo_dias[item] = ss_optimo_dias

# Convertir los diccionarios de resultados en DataFrames
resultados_stock_seguridad = pd.DataFrame({
    'Stock_Seguridad': stocks_seguridad,
    'Stock_Seguridad_D칤as': stocks_seguridad_dias,
    'Techo_Inventario_D칤as': techo_inventario_dias,
    'Stock_Seguridad_Optimo_D칤as': stocks_seguridad_optimo_dias
})

# Mostrar los resultados
print("\n游늵 Resultados de Stock de Seguridad:")
print(resultados_stock_seguridad.round(2))
```
